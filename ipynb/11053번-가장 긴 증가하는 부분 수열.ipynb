```python
n = int(input())
arr = list(map(int,input().split()))
answer = [1 for _ in range(n)]


for i in range(1,n):
    temp = answer[i]
    for j in range(0,i):
        if arr[j]<arr[i]:
            temp = max(temp, answer[j]+1)
    answer[i] = temp
    #answer[i]=max(answer[i-2],answer[i-1])
print(max(answer))
```
