```python
from collections import deque

def solution(board, r, c):
    answer = 0
    start = (r,c)
    end = board[r][c]
    _sum = 0
    #while not check(board):
    if board[r][c]==0:
        cnt, current = BFS1(board,start)
        _sum+=cnt
        y, x = current
        #print('BFS1')
        #print(cnt,current)
        start = (y,x)
        end = board[start[0]][start[1]]
        
    while True:
        #print(board)
        
        cnt, current = BFS2(board,start,end)#스타트 엔드 누르고 이동까지
        
        _sum+=cnt
        y, x = current
        #print('BFS2')
        #print(cnt,current)
        if check(board):
            break
        cnt, current = BFS1(board,current)
        _sum+=cnt
        y, x = current
        #print('BFS1')
        #print(cnt,current)
        start = (y,x)
        end = board[start[0]][start[1]]
        
    answer = _sum
    return answer

def check(board):
    for i in range(4):
        for j in range(4):
            if board[i][j]!=0:
                return False
    return True
def BFS1(board,start):
    coordinate = ()
    result = []
    dy = [0,1,-1,0]
    dx = [1,0,0,-1]
    q = deque()
    start_y,start_x = start
    q.append((start_y,start_x,[]))
    visit = set()
    visit.add(start)
    while q:
        cur_y,cur_x,log = q.popleft()
        #print(cur_y,cur_x,log)
        if board[cur_y][cur_x]!=0 and len(log)!=0:
            coordinate = (cur_y,cur_x)
            result.append((coordinate,log))
            #visit.remove(coordinate)
            #continue
            
        for i in range(4):
            ny = cur_y + dy[i]
            nx = cur_x + dx[i]
            if ny<4 and ny>=0 and nx<4 and nx>=0:
                if (ny,nx) not in visit:
                    visit.add((ny,nx))
                    if i==0:
                        log.append('right')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
                    elif i==1:
                        log.append('down')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
                    elif i==2:
                        log.append('up')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
                    elif i==3:
                        log.append('left')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
                        
    #print(result)
    pre_count=9999
    for re in result:
        count = len(set(re[1]))
        if pre_count>count:
            pre_count = count
            coordinate = re[0]
    #print((pre_count,coordinate))
    #print((len(result),coordinate))
    return (pre_count,coordinate)

            
def BFS2(board,start,end):
    result = []
    coordinate = ()
    dy = [0,1,-1,0]
    dx = [1,0,0,-1]
    q = deque()
    start_y,start_x = start
    q.append((start_y,start_x,[]))
    visit = set()
    visit.add(start)
    board[start_y][start_x]=0
    while q:
        cur_y,cur_x,log = q.popleft()
        #print(cur_y,cur_x,log)
        #print(cur_y,cur_x,log)
        if board[cur_y][cur_x]==end and len(log)!=0:
            coordinate = (cur_y,cur_x)
            result.append((coordinate,log))
            #visit.remove(coordinate)
            #continue
        for i in range(4):
            ny = cur_y + dy[i]
            nx = cur_x + dx[i]
            if ny<4 and ny>=0 and nx<4 and nx>=0:
                if (ny,nx) not in visit:
                    visit.add((ny,nx))
                    if i==0:
                        log.append('right')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
                    elif i==1:
                        log.append('down')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
                    elif i==2:
                        log.append('up')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
                    elif i==3:
                        log.append('left')
                        _log = log.copy()
                        q.append((ny,nx,_log))
                        log.pop()
    #print(result)
    pre_count=9999
    for re in result:
        count = len(set(re[1]))
        if pre_count>count:
            pre_count = count
            coordinate = re[0]
    board[coordinate[0]][coordinate[1]]=0
    #print((pre_count+2,coordinate))
    return (pre_count+2,coordinate)
    
    
board = [[3,0,0,2],[0,0,1,0],[0,1,0,0],[2,0,0,3]]
r = 0
c = 1
print(solution(board,r,c))
```
