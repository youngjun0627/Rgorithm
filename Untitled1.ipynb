```python
A,B = map(int,input().split())
n,m = map(int,input().split())


def move(robot, op):
    x,y,direc = robot
    if op=='L':
        if direc=='N':
            direc='W'
        elif direc=='E':
            direc='N'
        elif direc=='S':
            direc='E'
        elif direc=='W':
            direc='S'
    elif op=='R':
        if direc=='N':
            direc='E'
        elif direc=='E':
            direc='S'
        elif direc=='S':
            direc='W'
        elif direc=='W':
            direc='N'
    elif op=='F':
        if direc=='N':
            y+=1
        elif direc=='E':
            x+=1
        elif direc=='S':
            y-=1
        elif direc=='W':
            x-=1
    return x,y,direc

def solution(operation):
    _str = operation
    idx = int(_str[0])
    op = _str[1]
    go = int(_str[2])
    
    if op=='F':
        while go>0:
            go-=1
            cur = robots[idx-1]
            x,y,direc = move(cur,op)
            #print(x,y,direc)
            if x<=0 or x>A or y<=0 or y>B:
                print('Robot {0} crashes into the wall'.format(idx))
                return False
            for i, robot in enumerate(robots):
                if i==idx-1:
                    continue
                cx,cy,_ = robot
                
                if x==cx and y==cy:
                    print('Robot {0} crashes into robot {1}'.format(idx,i+1))
                    return False
                
            robots[idx-1]=((x,y,direc))
        return True
    else:
        cur = robots[idx-1]
        
        for _ in range(go):
            x,y,direc = move(cur,op)
            cur = (x,y,direc)
        robots[idx-1]=cur
        return True
    
        
operations = []
robots = []
for _ in range(n):
    temp = list(input().split())
    x = int(temp[0])
    y = int(temp[1])
    direc = temp[2]
    robots.append((x,y,direc))
for _ in range(m):
    operations.append(list(input().split()))
check=True
for operation in operations:
    check = solution(operation)
    if not check:
        break
if check:
    print('OK')


```
