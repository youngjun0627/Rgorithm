```python
from collections import deque
from itertools import combinations
import copy



def bfs(starts,board,n,m):
    result = 0
    q = deque()
    visit=set()
    dy = [0,1,0,-1]
    dx = [1,0,-1,0]
    for start in starts:
        q.append(start)
        visit.add(start)
    while q:
        cy,cx = q.popleft()
        
        board[cy][cx]=2
        for i in range(4):
            ny = cy + dy[i]
            nx = cx + dx[i]
            if ny>=0 and ny< n and nx>=0 and nx<m and board[ny][nx]==0:
                if (ny,nx) not in visit:
                    visit.add((ny,nx))
                    q.append((ny,nx))
    #print(board)
    for b in board:
        result += b.count(0)
    return result

if __name__ == '__main__':
    n,m = map(int,input().split())
    board = [[0 for _ in range(m)] for _ in range(n)]
    starts = []
    cands = []
    for i in range(n):
        temp = list(map(int,input().split()))
        for j in range(m):
            if temp[j]==2:
                starts.append((i,j))
            if temp[j]==0:
                cands.append((i,j))
            board[i][j]=temp[j]
    answer = 0
    for cand in combinations(cands,3):
        _board = copy.deepcopy(board)
        for c in cand:
            _board[c[0]][c[1]]=1
        result = bfs(starts,_board,n,m)
        answer = max(answer,result)
        #print(result)
    print(answer)

```
