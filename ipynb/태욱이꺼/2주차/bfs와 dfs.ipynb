```python
from collections import deque

dy = [0,0,-1,1]
dx = [1,-1,0,0]

def bfs(start,board):
    q = deque()
    q.append(start)
    visit = set([start])
    while q:
        cy,cx = q.popleft()
        for i in range(4):
            ny = cy+dy[i]
            nx = cx+cx[i]
            if ny>=0 and ny<len(board) and nx>=0 and nx<len(board):
                if (ny,nx) not in visit:
                    q.append((ny,nx))
                    visit.add((ny,nx))
                    
def dfs(start,board):
    stack = deque()
    stack.append(start)
    visit = set([start])
    while stack:
        cy,cx = stack.pop()
        for i in range(4):
            ny = cy+dy[i]
            nx = cx+cx[i]
            if ny>=0 and ny<len(board) and nx>=0 and nx<len(board):
                if (ny,nx) not in visit:
                    stack.append((ny,nx))
                    visit.add((ny,nx))

visit = set()
def dfs2(start,board):
    if start in visit:
        return
    visit.add(start)
    cy,cx = start
    for i in range(4):
        ny = cy+dy[i]
        nx = cx+cx[i]
        dfs2(((ny,nx),board))
```
