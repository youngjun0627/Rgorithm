```python
from collections import deque

n = int(input())
k = int(input())
board = [[0 for _ in range(n)] for _ in range(n)]
for _ in range(k):
    r,c = map(int,input().split())
    board[r-1][c-1]=2
board[0][0]=1

l = int(input())
time_c = [0 for _ in range(1,10001)]
time_c[0]=1
for _ in range(l):
    x,c = map(str,input().split())
    if c=='L':
        time_c[int(x)] = -1
    else:
        time_c[int(x)] = 1
    
q = deque()
body = [(0,0)]
dy = [-1,0,1,0]
dx = [0,-1,0,1]
q.append((0,0,3)) # y,x,state
ct=0
while q:
    cy,cx,state = q.popleft()
    #print(cy,cx,state)
    ny = cy+dy[state]
    nx = cx+dx[state]
    
    if ny>=n or ny<0 or nx>=n or nx<0 or board[ny][nx]==1:
        break
    if board[ny][nx]==0:
        tail = body[0]
        body.pop(0)
        board[tail[0]][tail[1]]=0
    
        
    board[ny][nx]=1
    body.append((ny,nx))    
    
    next_state = state
    if time_c[ct+1] == -1:
        if state==3:
            next_state = 0
        else:
            next_state = state+1
    elif time_c[ct+1] == 1:
        if state==0:
            next_state = 3
        else:
            next_state = state-1
    q.append((ny,nx,next_state))
    ct+=1
print(ct+1)
```
