```python
from collections import deque

n = int(input())
board = []
for _ in range(n):
    board.append(list(map(int,input().split())))
    
q = deque()
def up(board, n):
    result = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if board[j][i]!=0:
                q.append(board[j][i])
        pre = 0
        temp = []
        while q:
            cur = q.popleft()
            if pre==0:
                pre=cur
                continue
            if pre!=cur:
                temp.append(pre)
                pre=cur
            else:
                pre+=cur
                temp.append(pre)
                pre=0
        if pre:
            temp.append(pre)
        while len(temp)<n:
            temp.append(0)
        for j in range(n):
            result[j][i] = temp[j]
    return result

def down(board, n):
    result = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n-1,-1,-1):
            if board[j][i]!=0:
                q.append(board[j][i])
        pre = 0
        temp = []
        while q:
            cur = q.popleft()
            if pre==0:
                pre=cur
                continue
            if pre!=cur:
                temp.append(pre)
                pre=cur
            else:
                pre+=cur
                temp.append(pre)
                pre=0
        if pre:
            temp.append(pre)
        while len(temp)<n:
            temp.append(0)
        for j in range(n-1,-1,-1):
            result[j][i] = temp[n-1-j]
    return result

def left(board, n):
    result = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            if board[i][j]!=0:
                q.append(board[i][j])
        pre = 0
        temp = []
        while q:
            cur = q.popleft()
            if pre==0:
                pre=cur
                continue
            if pre!=cur:
                temp.append(pre)
                pre=cur
            else:
                pre+=cur
                temp.append(pre)
                pre=0
        if pre:
            temp.append(pre)
        while len(temp)<n:
            temp.append(0)
        for j in range(n):
            result[i][j] = temp[j]
    return result

def right(board, n):
    result = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n-1,-1,-1):
            if board[i][j]!=0:
                q.append(board[i][j])
        pre = 0
        temp = []
        while q:
            cur = q.popleft()
            if pre==0:
                pre=cur
                continue
            if pre!=cur:
                temp.append(pre)
                pre=cur
            else:
                pre+=cur
                temp.append(pre)
                pre=0
        if pre:
            temp.append(pre)
        while len(temp)<n:
            temp.append(0)
        for j in range(n-1,-1,-1):
            result[i][n-1-j] = temp[j]
    return result
    
answer = 0
def dfs(cnt, board,n):
    global answer
    global count
    if cnt==5:
        for i in range(n):
            answer = max(answer, max(board[i]))
        return
    
    
    
    dfs(cnt+1, up(board,n), n,)
    dfs(cnt+1, down(board,n), n)
    dfs(cnt+1, left(board,n), n)
    dfs(cnt+1, right(board,n), n)
dfs(0,board,n)
print(answer)
```
