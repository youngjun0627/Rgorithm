```python
from collections import deque

def solution(board):
    answer = 9999999
    robot = ((1,1,1), (1,2,2))
    q = deque()
    end = (len(board),len(board))
    visit = set()
    visit.add((1,1,1,2))
    dy=[0,0,1,-1]
    dx=[-1,1,0,0]
    q.append((robot,0))
    while q:
        cur_robot, cnt = q.popleft()
        cur_robot1,cur_robot2 = cur_robot
        cur_y1,cur_x1,cur_c1 = cur_robot1
        cur_y2,cur_x2,cur_c2 = cur_robot2
        if (cur_y1==end[0] and cur_x1==end[1]) or (cur_y2==end[0] and cur_x2==end[1]):
            answer = min(answer,cnt)
        for _dy,_dx in zip(dy,dx):
            ny1 = cur_y1+_dy
            ny2 = cur_y2+_dy
            nx1 = cur_x1+_dx
            nx2 = cur_x2+_dx
            if check(nx1,nx2,ny1,ny2,board) and (ny1,nx1,ny2,nx2) not in visit:
                visit.add((ny1,nx1,ny2,nx2))
                q.append((((ny1,nx1,cur_c1),(ny2,nx2,cur_c2)),cnt+1))
        if (cur_c1==1 and cur_c2==2) or (cur_c1==2 and cur_c2==1):
            pos = check2(cur_x1,cur_x2,cur_y1,cur_y2,board,1)
            if pos:
                for p in pos:
                    nx1,ny1,nx2,ny2=p
                    if (ny1,nx1,ny2,nx2) not in visit:
                        visit.add((ny1,nx1,ny2,nx2))
                        q.append((((ny1,nx1,3),(ny2,nx2,4)),cnt+1))
        elif (cur_c1==3 and cur_c2==4) or (cur_c1==4 and cur_c2==3):
            pos = check2(cur_x1,cur_x2,cur_y1,cur_y2,board,2)
            if pos:
                for p in pos:
                    nx1,ny1,nx2,ny2=p
                    if (ny1,nx1,ny2,nx2) not in visit:
                        visit.add((ny1,nx1,ny2,nx2))
                        q.append((((ny1,nx1,1),(ny2,nx2,2)),cnt+1))
        
                                                                                                             
            
    return answer

def check(nx1,nx2,ny1,ny2,board):
    return (nx1<=len(board) and nx1>0) and (nx2<=len(board) and nx2>0) and (ny2<=len(board) and ny2>0) and (ny1<=len(board) and ny1>0) and board[ny2-1][nx2-1]==0 and board[ny1-1][nx1-1]==0
                                                                                                            
def check2(nx1,nx2,ny1,ny2,board,mode):
    result = None
    pos_list = []
    temp = None
    if mode==1:
        result = [(nx1,ny1,nx1,ny1+1),(nx1,ny1,nx1,ny1-1),(nx2,ny2+1,nx2,ny2),(nx2,ny2-1,nx2,ny2)]
        temp = [(ny2+1,nx2),(ny2-1,nx2),(ny1+1,nx1),(ny1-1,nx1)]
    elif mode==2:
        result = [(nx1,ny1,nx1-1,ny1),(nx1,ny1,nx1+1,ny1),(nx2-1,ny2,nx2,ny2),(nx2+1,ny2,nx2,ny2)]
        temp = [(ny2,nx2-1),(ny2,nx2+1),(ny1,nx1-1),(ny1,nx1+1)]
        
    for re,t in zip(result,temp):
        nx1,ny1,nx2,ny2=re
        ty,tx=t
        if check(nx1,nx2,ny1,ny2,board) and board[ty-1][tx-1]==0:
            pos_list.append(re)
    return pos_list
```
