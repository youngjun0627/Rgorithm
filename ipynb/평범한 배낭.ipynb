```python
n,k = map(int,input().split())
bag = []
for _ in range(n):
    w,v = map(int,input().split())
    bag.append((w,v))
board = [[0 for _ in range(k+1)] for _ in range(n+1)] # row = i, col = weight
for i in range(1,n+1):
    for j in range(k+1):
        cur_w, cur_v = bag[i-1]
        if j<cur_w:
            board[i][j] = board[i-1][j]
        else:
            if board[i-1][j-cur_w]+cur_v>board[i-1][j]:
                board[i][j] = board[i-1][j-cur_w]+cur_v
            else:
                board[i][j] = board[i-1][j]
                
print(board[n][k])
    
```
