```python
from collections import deque

def bfs(start, dic):
    q = deque()
    q.append(start)
    visit=[start]
    while q:
        cur = q.popleft()
        
        dic[cur].sort()
        if cur not in dic:
            continue
        for _next in dic[cur]:
            if _next not in visit:
                visit.append(_next)
                
                q.append(_next)
    return visit

def dfs(start,dic):
    stack = deque()
    stack.appendleft(start)
    visit=[start]
    while stack:
        cur = stack.popleft()
        if cur not in visit:
            visit.append(cur)
        dic[cur].sort(key=lambda x: -x)
        
        if cur not in dic:
            continue
        for _next in dic[cur]:
            if _next not in visit:
                
                stack.appendleft(_next)
                
        
    return visit
                
n,m,start = map(int,input().split())
dic = {}
for _ in range(m):
    a,b = map(int,input().split())
    if a not in dic:
        dic[a]=[]
    if b not in dic:
        dic[b] = []
        
    dic[a].append(b)
    dic[b].append(a)

    

df = dfs(start,dic)

bf = bfs(start,dic)

answer1 = ''
answer2 = ''
for idx,a in enumerate(df):
    if idx==len(df)-1:
        answer1+=str(a)
    else:
        answer1+=str(a)+' '
answer2 = ''
for idx,a in enumerate(bf):
    if idx==len(df)-1:
        answer2+=str(a)
    else:
        answer2+=str(a)+' '
print(answer1)
print(answer2)
```
