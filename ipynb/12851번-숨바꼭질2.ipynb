```python
from collections import deque

dp = [int(1e9)] * 100001
def bfs(start,end):
    global dp
    q = deque()
    q.append((start,0))
    visit = set()
    visit.add(start)
    while q:
        cur,cnt = q.popleft()
        visit.add(cur)
        if dp[cur]<cnt:
            continue
        dp[cur]=cnt
        if cur==end:
            break
        if cur+1<=100000 and cur+1 not in visit:
            q.append((cur+1,cnt+1))
        if cur*2<=100000 and cur*2 not in visit:
            q.append((cur*2,cnt+1))
        if cur-1>=0 and cur-1 not in visit:
            q.append((cur-1,cnt+1))

                    
def check(start,end, n):
    q = deque()
    q.append((start,0))
    visit = set()
    visit.add(start)
    result = 0
    while q:
        cur,cnt = q.popleft()
        visit.add(cur)
        if cnt>n:
            continue
        if cur==end:
            result += 1
        if cur+1<=100000 and cur+1 not in visit:
            q.append((cur+1,cnt+1))
        if cur*2<=100000 and cur*2 not in visit:
            q.append((cur*2,cnt+1))
        if cur-1>=0 and cur-1 not in visit:
            q.append((cur-1,cnt+1))
    return result
a,b = map(int,input().split())
bfs(a,b)
n = dp[b]
print(n)
print(check(a,b,n))
```
