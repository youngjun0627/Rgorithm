```python
n = int(input())

board = [[0 for _ in range(n)] for _ in range(n)]

for i in range(n):
    temp = list(map(int,input().split()))
    for j in range(n):
        board[i][j] = temp[j]
        
dp = [[[0 for _ in range(n)] for _ in range(n)] for _ in range(3)]

dp[0][0][1] = 1

for i in range(2,n):
    if board[0][i]==0:
        dp[0][0][i]=dp[0][0][i-1]

    
for i in range(1,n):
    for j in range(1,n):
        if board[i][j]==0 and board[i-1][j]==0 and board[i][j-1]==0:
            dp[2][i][j] = dp[1][i-1][j-1] + dp[0][i-1][j-1] + dp[2][i-1][j-1]
        
        if board[i][j]==0:
            dp[0][i][j] = dp[0][i][j-1] + dp[2][i][j-1]
            dp[1][i][j] = dp[1][i-1][j] + dp[2][i-1][j]
print(dp)
print(sum(dp[i][n-1][n-1] for i in range(3)))
```

```python
### 시간 초과 코드 ###
import sys

sys.setrecursionlimit(300)

def dfs(start,board,state,n):
    global answer
    #print(start)
    if start[0]+1==n and start[1]+1==n:
        answer+=1
        return
    
    dy = [0,1,1]
    dx = [1,0,1]
    
    for i in range(3):
        ny = start[0]+dy[i]
        nx = start[1]+dx[i]
        
        if ny>=0 and ny<n and nx>=0 and nx<n and board[ny][nx]==0:
            if i==0:
                if state==2 or state==0:
                    dfs((ny,nx),board,0,n)
            if i==1:
                if state==2 or state==1:
                    dfs((ny,nx),board,1,n)
            if i==2:
                if board[ny-1][nx]!=1 and board[ny][nx-1]!=1:
                    dfs((ny,nx),board,2,n)
            
    return

n = int(input())

board = [[0 for _ in range(n)] for _ in range(n)]

for i in range(n):
    temp = list(map(int,input().split()))
    for j in range(n):
        board[i][j] = temp[j]
        
start = (0,1)

answer = 0
dfs(start,board,0,n)

print(answer)
    
```
