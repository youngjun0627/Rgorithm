```python
# 카카오 블라인드 채용 5번

# n은 벽면 크기
# build_frame -> x,y,a,b x,y좌표 a는 0이면 기둥 1은 보, b는 0은 삭제 1은 설치

def check_wall(walls,boe,curr):
    current_x,current_y = curr
    if current_y==0:
        return True
    for b in boe:
        b_x,b_y = b
        if (current_x==b_x or current_x==b_x+1) and current_y==b_y:
            return True
    for w in walls:
        w_x,w_y = w
        if current_x==w_x and current_y==w_y+1:
            return True
    return False


def check_boe(walls,boe,curr):
    current_x,current_y = curr
    if (current_x-1,current_y) in boe and (current_x+1,current_y) in boe:
        return True
    for w in walls:
        w_x,w_y = w
        if (current_x==w_x or current_x+1 == w_x) and current_y==w_y+1:
            return True
    return False



def solution(n, build_frame):
    answer = []
    walls = set()
    boe = set()
    for x,y,a,b in build_frame:
        if a==1:#보
            if b==1:
                if check_boe(walls,boe,(x,y)):
                    boe.add((x,y))
            else:
                if (x,y) not in boe:
                    continue
                boe.remove((x,y))
                for wall in walls:
                    if not check_wall(walls,boe,wall):
                        boe.add((x,y))
                        break
                for bo in boe:
                    if not check_boe(walls,boe,bo):
                        boe.add((x,y))
                        break
                
        else:
            if b==1:
                if check_wall(walls,boe,(x,y)):
                    walls.add((x,y))
            else:
                if (x,y) not in walls:
                    continue
                walls.remove((x,y))
                for wall in walls:
                    if not check_wall(walls,boe,wall):
                        walls.add((x,y))
                        break
                for bo in boe:
                    if not check_boe(walls,boe,bo):
                        walls.add((x,y))
                        break
        
    for w in walls:
        a,b = w
        answer.append([a,b,0])
    for bo in boe:
        a,b = bo
        answer.append([a,b,1])
    answer.sort()
    return answer

build_frame = [[0,0,0,1],[2,0,0,1],[4,0,0,1],[0,1,1,1],[1,1,1,1],[2,1,1,1],[3,1,1,1],[2,0,0,0],[1,1,1,0],[2,2,0,1]]
n=5
solution(n,build_frame)
```

```python
def checkwall(cur, walls, boes):
    cur_y,cur_x = cur
    if cur_y==0:
        return True
    for wall in walls:
        com_y,com_x,_ = wall
        if com_y+1==cur_y and com_x==cur_x:
            return True
    for boe in boes:
        com_y, com_x,_ = boe
        if com_y==cur_y and (com_x==cur_x or com_x+1==cur_x):
            return True
    return False

def checkboe(cur, boes, walls):
    cur_y,cur_x = cur
    for wall in walls:
        com_y, com_x,_ = wall
        if com_y+1==cur_y and (com_x==cur_x or com_x==cur_x+1):
            return True
    check1=False
    check2=False
    for boe in boes:
        com_y, com_x,_ = boe
        if com_y==cur_y and cur_x==com_x+1:
            check1=True
        if com_y==cur_y and cur_x==com_x-1:
            check2=True
    if check1 and check2:
        return True
    return False

def solution(n,build_frame):
    answer = []
    walls = []
    boes = []
    for bf in build_frame:
        x,y,n,m = bf
        if n==0: # 기둥
            if m==1: # 설치
                if checkboe((y,x),boes,walls) and checkwall((y,x),walls,boes):
                    walls.append((y,x,0))
            else if m==0: # 제거
                walls.remove((y,x,0))
        else if n==1: # 보
            if m==1: # 설치
                if checkboe((y,x),boes,walls) and checkwall((y,x),walls,boes):
                    boes.append((y,x,1))
            else if m==0: # 제거
                boes.remove((y,x,1))
    answer = walls+boes
    answer.sort(key = lambda x: x[0],x[1],-x[2])
    return answer
```
