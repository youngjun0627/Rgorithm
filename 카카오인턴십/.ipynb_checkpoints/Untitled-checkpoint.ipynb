```python
# 1번
def solution(s):
    answer = ''
    stack=''
    dic = {'zero':0, 'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':9}
    
    for i in s:
        if i.isdigit():
            answer+=i
            
        else:
            stack+=i
            if stack in dic.keys():
                answer+=str(dic[stack])
                stack=''
    return int(answer)


s = "one4seveneight"
solution(s)
```

```python
# 2번
from collections import deque

def bfs(start,board,people):
    q = deque()
    cnt=0
    visit = set()
    q.append((0,start[0],start[1]))
    dy = [0,0,-1,1]
    dx = [-1,1,0,0]
    visit.add(start)
    result=-1
    while q:
        cnt,y,x = q.popleft()
        if (y,x) in people and (y,x) != start:
            result = cnt
            break
            
        for i in range(4):
            ny=y+dy[i]
            nx=x+dx[i]
            if 0<=ny<5 and 0<=nx<5 and board[ny][nx]!='X':
                if (ny,nx) not in visit:
                    visit.add((ny,nx))
                    q.append((cnt+1,ny,nx))
    
    if result>2 or result==-1:
        return True
    else:
        return False
    
def solution(places):
    answer = []
    for k in range(5):
        people = []
        board = [['' for _ in range(5)] for _ in range(5)]
        for i in range(5):
            for j in range(5):
                if places[k][i][j]=='P':
                    people.append((i,j))
                board[i][j] = places[k][i][j]
        check=True
        for person in people:
            if bfs(person,board,people):
                continue
            else:
                check=False
                break
        if check:
            answer.append(1)
        else:
            answer.append(0)
        
    return answer
places = [["POOOP", "OXXOX", "OPXPX", "OOXOX", "POXXP"], ["POOPX", "OXPXP", "PXXXO", "OXXXO", "OOOPP"], ["PXOPX", "OXOXP", "OXPXX", "OXXXP", "POOXX"], ["OOOXX", "XOOOX", "OOOXX", "OXOOX", "OOOOO"], ["PXPXP", "XPXPX", "PXPXP", "XPXPX", "PXPXP"]]
solution(places)
```

```python
from collections import deque
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self, data):
        self.head = Node(data)
    
    #헤더부터 탐색해 뒤에 새로운 노드 추가하기
    def append(self, data):
        cur = self.head
        while cur.next is not None:
            cur = cur.next
        cur.next = Node(data)

    #모든 노드 값 출력
    def print_all(self):
        cur = self.head
        result = ''
        idx = 0
        while cur is not None:
            data = cur.data
            if idx!=data:
                result+='X'
            else:
                result+='O'
                cur = cur.next
            idx+=1
        return result
    def get_node(self, index):
        cnt = 0
        node = self.head
        while cnt < index:
            cnt += 1
            node = node.next
        return node
    
    def add_node(self, index, value):
        new_node = Node(value)
        if index == 0:
            new_node.next = self.head
            self.head = new_node
            return
        node = self.get_node(index-1)
        next_node = node.next
        node.next = new_node
        new_node.next = next_node
        
    def delete_node(self, index):
        if index == 0:
            self.head = self.head.next
            return
        node = self.get_node(index-1)
        node.next = node.next.next

def solution(n, k, cmd):
    answer = ''
    stack = deque()
    linkedlist = LinkedList(0)
    for i in range(1,n):
        linkedlist.append(i)
        
    for idx,c in enumerate(cmd):
        if c.startswith('U'):
            k-=int(c.split()[1])
        elif c.startswith('D'):
            k+=int(c.split()[1])
        elif c.startswith('C'):
            
            stack.appendleft((k,linkedlist.get_node(k).data))
            linkedlist.delete_node(k)
            if k==n-1:
                k-=1
            n-=1
        elif c.startswith('Z'):
            cur,value = stack.popleft()
            linkedlist.add_node(cur,value)
            if cur<=k:
                k+=1
            n+=1
    answer = linkedlist.print_all()
    return answer
n = 8
k = 2
cmd = ["D 2","C","U 3","C","D 4","C","U 2","Z","Z","U 1","C"]
solution(n,k,cmd)
```

```python
from collections import deque
    
def solution(n, k, cmd):
    answer = ''
    arr = [i for i in range(n)]
    cur = k
    remove_list = deque()
    for command in cmd:
        if command.startswith('U'):
            cur-=int(command[2:])
        elif command.startswith('D'):
            cur+=int(command[2:])
        elif command=='C':
            remove_list.appendleft((cur,arr[cur]))
            arr.pop(cur)
            if cur==len(arr):
                cur-=1
            
            
        elif command=='Z':
            remove_index,remove_value = remove_list.popleft()
            if remove_index<cur:
                cur+=1
            arr.insert(remove_index,remove_value)
        print(cur,arr)
    for remove,_ in remove_list:
        arr.insert(remove,-1)
    
    for key,value in enumerate(arr):
        if key==value:
            answer+='O'
        else:
            answer+='X'
    return answer

n=8
k=2
cmd = ["D 2","C","U 3","C","D 4","C","U 2","Z","Z","U 1","C"]
solution(n,k,cmd)
```
