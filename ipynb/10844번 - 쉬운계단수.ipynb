```python
dp = [[0 for _ in range(10)] for _ in range(101)]
for i in range(1,10):
    dp[0][i]=1

for i in range(1,101):
    for j in range(10):
        if j==0:
            dp[i][j]=dp[i-1][j+1]
        elif j==9:
            dp[i][j] = dp[i-1][j-1]
        else:
            dp[i][j] = (dp[i-1][j-1]+dp[i-1][j+1])%1000000000
n = int(input())
print(sum(dp[n-1])%1000000000)
```
