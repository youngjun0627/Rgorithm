```python
def solution(id):
    result = check1(id)
    print(result)
    result = check2(result)
    print(result)
    result = check3(result)
    print(result)
    result = check4(result)
    print(result)
    result = check5(result)
    print(result)
    result = check6(result)
    print(result)
    result = check7(result)
    print(result)
    answer = result
    return answer

def check1(id):
    result = id.lower()
    return result

def check2(id):
    check_list = ['-','_','.','a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x,','y','z','1','2','3','4','5','6','7','8','9','0']
    result = []
    for i in id:
        if i not in check_list:
            continue
        result.append(i)
    return ''.join(result)

def check3(id):
    cnt = 0
    result = []
    for i in id:
        if i=='.':
            cnt+=1
            if cnt>1:
                continue
            else:
                result.append(i)
        else:
            cnt=0
            result.append(i)
    return ''.join(result)

def check4(id):
    check1 = False
    check2 = False
    if id[0]=='.':
        check1 = True
    if id[len(id)-1]=='.':
        check2 = True
    
    if check1 and not check2:
        return id[1:]
    elif not check1 and check2:
        return id[:len(id)-1]
    elif check1 and check2:
        return id[1:len(id)-1]
    else:
        return id
        
def check5(id):
    if id=='':
        return 'a'
    else:
        return id

def check6(id):
    result = ''
    if len(id)>=16:
        result = id[:15]
        result = check4(result)
    else:
        result = id
    return result

def check7(id):
    result = ''
    if len(id)==0:
        result = 'aaa'
    elif len(id)==1:
        temp = [id[0],id[0],id[0]]
        result = ''.join(temp)
    elif len(id)==2:
        temp = [id[0],id[1],id[1]]
        result = ''.join(temp)
    else:
        result = id
    return result
    

id = "sdfs.dr322.41234as.d.asdasd...as.dasdfasf325q...sdaf..."

solution(id)
```
