```python
def change(time):
    a,b,c = time.split()
    h,m,s = b.split(':')
    h=int(h)
    m = int(m)
    s = float(s)
    end = h*60*60 + m*60 + s
    c = float(c[:-1])
    start = end-c+0.001
    return start, end
def solution(lines):
    answer = 0
    times = []
    for line in lines:
        times.append(change(line))
    for p1,p2 in times:
        for p in [p1,p2]:
            cnt=0
            for start, end in times:
                if p<=start<p+1:
                    cnt+=1
                elif p<=end<p+1:
                    cnt+=1
                elif start<=p and p+1<=end:
                    cnt+=1
            answer = max(answer, cnt)
    return answer
```
