```python
from queue import PriorityQueue

def dijk(start,dic,n):
    arr = [99999999 for _ in range(n)]
    arr[start-1]=0
    pq = PriorityQueue()
    pq.put((0,start))
    visit=set()
    visit.add(start)
    while not pq.empty():
        if len(visit)==n:
            break
        cost,start = pq.get()
        visit.add(start)
        for _next in dic[start]:
            if _next not in visit:
                if arr[_next-1] > cost+dic[start][_next]:
                    arr[_next-1]=cost+dic[start][_next]
                    pq.put((arr[_next-1],_next))
    return arr

n,e = map(int,input().split())
graph = {}
start = 1
end = n
for i in range(1,n+1):
    graph[i]={}
for _ in range(e):
    a,b,c = map(int,input().split())
    graph[a][b]=c
    graph[b][a]=c
v1,v2 =  map(int,input().split())
tasks = [(start,v1,v2,end),(start,v2,v1,end)]
answer = 99999999
for task in tasks:
    start, v1, v2, end = task
    arr1 = dijk(start,graph,n)
    arr2 = dijk(v1,graph,n)
    arr3 = dijk(v2,graph,n)
    answer = min(answer,arr1[v1-1]+arr2[v2-1]+arr3[end-1])
if answer==99999999:
    print(-1)
else:
    print(answer)
```
